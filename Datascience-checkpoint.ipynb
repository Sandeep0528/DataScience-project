```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"creditcard.csv")
dataset.head(3)
```

```python
dataset["Class"].value_counts
```

```python
legit=dataset[dataset.Class==0]
fraud=dataset[dataset.Class==1]
```

```python
len(legit),len(fraud)
```

```python
print(fraud.shape[0])
```

```python
train_legit_sample = legit.sample(n=300, random_state=42)
train_fraud_sample = fraud.sample(n=285, random_state=42)
test_legit_sample = legit.sample(n=200)
test_fraud_sample = fraud.sample(n=200)
```

```python
new_dataset=pd.concat([train_legit_sample,train_fraud_sample],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset["Class"].value_counts()
```

```python
x=new_dataset.iloc[:,:-1]
y=new_dataset["Class"]
x,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
lr.score(x_train,y_train)*100,lr.score(x_test,y_test)*100
```

```python
predictions=lr.predict(new_dataset.iloc[:,:-1])
new_dataset["Predictions"]=predictions
```
